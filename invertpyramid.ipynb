```python
def print_pyramid(n):
    for i in range(1, n + 1):
        # Print leading spaces
        for j in range(n - i):
            print(" ", end="")

        # Print stars
        for j in range(2 * i - 1):
            print("*", end="")

        # Move to the next line
        print()

# Input from user for pyramid height
height = int(input("Enter the height of the pyramid: "))
print_pyramid(height)

```

```python
def print_inverted_pyramid(n):
    for i in range(n, 0, -1):
        # Print leading spaces
        for j in range(n - i):
            print(" ", end="")

        # Print stars
        for j in range(2 * i - 1):
            print("*", end="")

        # Move to the next line
        print()

# Input from user for inverted pyramid height
height = int(input("Enter the height of the inverted pyramid: "))
print_inverted_pyramid(height)
```
